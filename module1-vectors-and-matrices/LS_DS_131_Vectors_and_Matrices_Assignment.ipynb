## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = [1,1,1,8]
h = [3,3,3,3]
```

```
def mag(x): 
    return math.sqrt(sum(i**2 for i in x))
```

```
mag(g)
```

```
mag(h)
```

- The magitude of g is longer than h
